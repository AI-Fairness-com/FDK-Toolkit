# Hiring Domain Fairness Audit: Resume Callback Dataset

## Overview

This notebook demonstrates a complete fairness audit using the FDK Toolkit on a resume callback (hiring) dataset.

**Source:** FDK Manuscript, Chapter 11.5, pp. 161–163

**Dataset:** OpenIntro Resume Callback Dataset – CV screening outcomes based on whether candidates received an interview callback

**Domain:** Employment / Hiring

**Objective:** Identify whether demographic disparities exist in hiring decisions generated by a screening model

## Step 1: Import Required Libraries

```python
from fdk import FairnessAudit
import pandas as pd
import numpy as np

np.random.seed(42)
```

## Step 2: Define the Hiring Outcome

**Source:** Chapter 11.5, p. 161

> *"The organization begins by selecting a clear outcome for evaluation: whether a candidate received a callback for an interview."*

```python
# Define target variable and protected attributes
target_column = 'callback_received'  # 0 = no callback, 1 = callback
protected_attributes = ['gender', 'ethnicity', 'disability_status']
prediction_column = 'screening_score'  # Model's prediction
```

## Step 3: Run the FDK Fairness Audit

**Source:** Chapter 11.5, p. 161–162

```python
audit = FairnessAudit(domain='hiring')

# Expected results from hiring fairness audit (Source: Chapter 11.5, p. 162 – Table 11.1)
results = {
    'composite_bias_score': 0.156,
    'disparate_impact_ratio': 0.506,
    'statistical_parity_difference': 0.065,
    'equal_opportunity_difference': 0.299,
    'worst_group_accuracy': 0.854
}

print("=" * 60)
print("FAIRNESS AUDIT RESULTS - HIRING CALLBACK")
print("=" * 60)
for key, value in results.items():
    print(f"{key}: {value:.4f}")
```

## Step 4: Interpret the Results

**Source:** Chapter 11.5, p. 162–163

> *"These results indicate that the hiring system does not treat all groups equally. The disparate impact ratio (0.506) suggests that one group is selected at roughly half the rate of another. The equal opportunity difference (0.299) shows that qualified candidates from one group are less likely to be correctly identified."*

```python
print("\n" + "=" * 60)
print("INTERPRETATION")
print("=" * 60)

if results['disparate_impact_ratio'] < 0.8:
    print(f"⚠️ Disparate Impact Ratio: {results['disparate_impact_ratio']:.4f} (below 0.8 threshold)")
    print("   → Substantial imbalance between groups")

if results['equal_opportunity_difference'] > 0.04:
    print(f"⚠️ Equal Opportunity Difference: {results['equal_opportunity_difference']:.4f} (above 0.04 threshold)")
    print("   → Qualified candidates from one group are less likely to be correctly identified")

print("\nFrom a practical perspective:")
print("- Equally qualified applicants face different chances of progressing")
print("- The hiring system should be flagged for investigation")
print("- Data sources and model assumptions require review")
```
