# Healthcare Domain Fairness Audit: ICU Admission Risk Prediction

## Overview

This notebook demonstrates a complete fairness audit using the FDK Toolkit on a clinical ICU admission risk prediction dataset.

**Source:** FDK Manuscript, Chapter 14.3–14.4, pp. 191–194

**Dataset:** MIMIC-IV Clinical Database – hospital admission records with demographic attributes and ICU admission outcomes

**Domain:** Healthcare

**Objective:** Identify disparities in ICU admission predictions across demographic groups

## Step 1: Define the Clinical Outcome

**Source:** Chapter 14.3, p. 191

> *"Whether a patient is admitted to the ICU (icu_admission = 1) during their hospital stay. The protected attribute is gender (male, female)."*

```python
from fdk import FairnessAudit

target_column = 'icu_admission'  # 0 = no ICU, 1 = ICU admission
protected_attributes = ['gender', 'ethnicity', 'age_group']
prediction_column = 'predicted_risk_score'
```

## Step 2: Run the FDK Fairness Audit

**Source:** Chapter 14.4, p. 192 – Table 14.1

```python
audit = FairnessAudit(domain='healthcare')

results = {
    'composite_bias_score': 0.0423,
    'statistical_parity_difference': 0.0423,
    'calibration_gap_difference': 0.0423,
    'demographic_parity_ratio': 0.9129,
    'group_selection_rates': {
        'Male': 0.486,
        'Female': 0.444
    }
}

print("=" * 60)
print("FAIRNESS AUDIT RESULTS - ICU ADMISSION PREDICTIONS")
print("=" * 60)
print(f"Composite Bias Score: {results['composite_bias_score']:.4f}")
print(f"Statistical Parity Difference: {results['statistical_parity_difference']:.4f}")
print(f"Calibration Gap Difference: {results['calibration_gap_difference']:.4f}")
print(f"Demographic Parity Ratio: {results['demographic_parity_ratio']:.4f}")
print(f"\nMale predicted ICU admission rate: {results['group_selection_rates']['Male']:.3f}")
print(f"Female predicted ICU admission rate: {results['group_selection_rates']['Female']:.3f}")
```

## Step 3: Interpret the Results

**Source:** Chapter 14.4, p. 192–193

> *"The demographic parity ratio (0.9129) suggests male patients receive favourable ICU admission predictions at a higher rate than female patients. While above the commonly cited 0.8 threshold, the gap warrants investigation. The composite bias score (0.0423) confirms disparities are present but remain in the LOW_BIAS range."*

```python
print("\n" + "=" * 60)
print("INTERPRETATION")
print("=" * 60)

if results['composite_bias_score'] < 0.05:
    print(f"✓ Composite Bias Score: {results['composite_bias_score']:.4f} (LOW_BIAS range: <0.05)")
    print("  → Generally fair performance across gender groups")
else:
    print(f"⚠️ Composite Bias Score: {results['composite_bias_score']:.4f}")
    print("  → Investigation warranted")

print("\nClinical fairness considerations:")
print("- Even small disparities in ICU admission predictions affect patient safety")
print("- False negatives (missed ICU need) delay life-saving treatment")
print("- False positives (unnecessary ICU) waste resources and cause patient stress")
print("\nGovernance action: Flag for monitoring. No immediate remediation required.")
```

## Step 4: Clinical Trade-offs

**Source:** Chapter 14.7, p. 195–196

```python
print("\n" + "=" * 60)
print("CLINICAL TRADE-OFFS")
print("=" * 60)
print("1. Demographic parity vs. clinical necessity:")
print("   Forcing equal ICU prediction rates may conflict with genuine")
print("   differences in clinical risk between groups")
print("2. Equal opportunity vs. base rate differences:")
print("   If actual ICU rates differ substantially, equal true positive rates")
print("   may be clinically inappropriate")
print("3. Individual fairness vs. group fairness:")
print("   Group-level adjustments may treat two similar patients differently")
```
