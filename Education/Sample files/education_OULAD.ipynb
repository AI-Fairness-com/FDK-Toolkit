# Education Domain Fairness Audit: Student Success Prediction

## Overview

This notebook demonstrates a complete fairness audit using the FDK Toolkit on a student success prediction dataset.

**Source:** FDK Manuscript, Chapter 12.3, pp. 168–169

**Dataset:** Open University Learning Analytics Dataset (OULAD) – student records and academic outcome predictions

**Domain:** Education

**Objective:** Determine whether student success predictions are fair across demographic groups

## Step 1: Define the Educational Outcome

**Source:** Chapter 12.3, p. 168

> *"Whether a student is predicted to achieve academic success (pass or distinction) versus failure or withdrawal."*

```python
from fdk import FairnessAudit

target_column = 'academic_success'  # 0 = fail/withdraw, 1 = pass/distinction
protected_attributes = ['gender', 'disability', 'region']
prediction_column = 'predicted_grade'
```

## Step 2: Run the FDK Fairness Audit

**Source:** Chapter 12.3, p. 169 – Table 12.1

```python
audit = FairnessAudit(domain='education')

results = {
    'composite_fairness_score': 0.7128,
    'disparate_impact_ratio': 0.9162,
    'statistical_parity_difference': 0.0386,
    'equal_opportunity_difference': 0.0339,
    'predictive_parity_difference': 0.0121,
    'academic_calibration_gap': 0.0211
}

print("=" * 60)
print("FAIRNESS AUDIT RESULTS - STUDENT SUCCESS")
print("=" * 60)
for key, value in results.items():
    print(f"{key}: {value:.4f}")
```

## Step 3: Interpret the Results

**Source:** Chapter 12.3, p. 169

> *"The disparate impact ratio (0.9162) indicates one group is predicted to succeed at a slightly lower rate than another. The equal opportunity difference (0.0339) shows that academically similar students are not always equally identified as successful."*

```python
print("\n" + "=" * 60)
print("INTERPRETATION")
print("=" * 60)

if results['disparate_impact_ratio'] < 0.85:
    print(f"⚠️ Disparate Impact Ratio: {results['disparate_impact_ratio']:.4f}")
    print("   → In education, even small disparities (<0.85) warrant investigation")
    print("   → Cumulative effects over years magnify initial differences")

print("\nStudents from different demographic backgrounds may receive")
print("systematically different predictions, potentially influencing")
print("tutoring or remediation allocation.")
```
